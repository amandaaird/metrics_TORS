```python
# SLIM
Sep_FAR = pd.read_csv('Kiva/SLIM/Sep/FAR/Lowest_Item.csv', header=None, names=['Sep_FAR'])
Sep_PFAR = pd.read_csv('Kiva/SLIM/Sep/PFAR/Lowest_Item.csv', header=None, names=['Sep_PFAR'])
Sep_Fairstar = pd.read_csv('Kiva/SLIM/Sep/Fairstar/Lowest_Item.csv', header=None, names=['Sep_Fairstar'])

JOa_FAR = pd.read_csv('Kiva/SLIM/JOa/FAR/Lowest_Item.csv', header=None, names=['Sep_FAR'])
JOa_PFAR = pd.read_csv('Kiva/SLIM/JOa/PFAR/Lowest_Item.csv', header=None, names=['Sep_PFAR'])
JOa_Fairstar = pd.read_csv('Kiva/SLIM/JOa/Fairstar/Lowest_Item.csv', header=None, names=['Sep_Fairstar'])


Sep_FAR = Sep_FAR.loc[~(Sep_FAR==-1).all(axis=1)]
Sep_PFAR = Sep_PFAR.loc[~(Sep_PFAR==-1).all(axis=1)]
Sep_Fairstar = Sep_Fairstar.loc[~(Sep_Fairstar==-1).all(axis=1)]

JOa_FAR = JOa_FAR.loc[~(JOa_FAR==-1).all(axis=1)]
JOa_PFAR = JOa_PFAR.loc[~(JOa_PFAR==-1).all(axis=1)]
JOa_Fairstar = JOa_Fairstar.loc[~(JOa_Fairstar==-1).all(axis=1)]
```

```python
#Just separate and Joint additive
fig, axs = plt.subplots(2, 3)
plt.setp(axs, xlim=(0,50), ylim=(0,4000))
fig.suptitle('SLIM')

axs[0, 0].hist(Sep_FAR)
axs[0, 0].set_title('FAR')
axs[0, 0].set(ylabel='Separate')
axs[0, 1].hist(Sep_PFAR)
axs[0, 1].set_title('PFAR')
axs[0, 2].hist(Sep_Fairstar)
axs[0, 2].set_title('Fairstar')

axs[1, 0].hist(JOa_FAR)
axs[1, 0].set(ylabel='Joint Additive')
axs[1, 1].hist(JOa_PFAR)
axs[1, 2].hist(JOa_Fairstar)


plt.tight_layout()

#for ax in axs.flat:
 #   ax.set(xlabel='Reranker', ylabel='Optimization Type')

#Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
# Full Plot
fig, axs = plt.subplots(3, 3)
plt.setp(axs, xlim=(0,50), ylim=(0,4000))
fig.suptitle('BPR')

axs[0, 0].hist(Sep_FAR)
axs[0, 0].set_title('FAR')
axs[0, 0].set(ylabel='Separate')
axs[0, 1].hist(Sep_PFAR)
axs[0, 1].set_title('PFAR')
axs[0, 2].hist(Sep_Fairstar)
axs[0, 2].set_title('Fairstar')

axs[1, 0].hist(JOa_FAR)
axs[1, 0].set(ylabel='Joint Additive')
axs[1, 1].hist(JOa_PFAR)
axs[1, 2].hist(JOa_Fairstar)

axs[2, 0].hist(JOm_FAR)
axs[2, 0].set(ylabel = 'Joint Multiplicative')
axs[2, 1].hist(JOm_PFAR)
axs[2, 2].hist(JOm_Fairstar)

plt.tight_layout()

#for ax in axs.flat:
 #   ax.set(xlabel='Reranker', ylabel='Optimization Type')

#Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
